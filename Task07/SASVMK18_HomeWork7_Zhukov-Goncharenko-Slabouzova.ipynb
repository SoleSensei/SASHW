### Задание

В компьютерном сервисном центре работают четыре сотрудника, специализирующихся на ремонте трех брендов дисководов для стационарных компьютеров. Сервисный центр хочет исследовать влияние специалиста и бренда дисковода на  
время обслуживания. В таблице disks  содержатся следующие данные:

 - Technician - имя специалиста (Angela, Bob, Justin, Karen)
 - Brand 	  - бренд дисковода (1, 2, или 3)
 - Time	      - время обслуживания (в минутах)
 
**Внимание:** Отвечая на вопросы ниже, пожалуйста, указывайте на _что_ в результатах работы вы смотрели и почему сделали такой вывод. 

#### 1. 
Проведите двухфакторный дисперсионный анализ, где Time - зависимая переменная, а Technician и Brand - независимые. Исследуйте пересечение независимых переменных в своей модели. 

#### Q1: Можно ли доверять результатам анализа? Почему?
Чтобы проверить корректность результатов анализа необходимо проверить выполнение условий two-way anova test.

1) Зависимая переменная `Time` типа continuous

2) Есть две независимые переменные `Brand` и `Technician` типа categorical

3) Данные были получены независимо друг от друга

4) Дисперсия данных совпадает

5) Распределение данных нормальное

Выполнение первых трех условий очевидно. Выполнение последних двух нет, для их проверки нужно проводить дополнительные тесты. Но если предположить, что они выполняются, то результатам анализа доверять можно.

#### Q2: При условии, что уровень значимости равен 0.05, является ли значимым общий F-тест для модели? Почему?
Да, является, потому что `Pr>F` < 0.05 (`Pr>F` берется из построенной таблицы для строки `Model`)

#### Q3: Стоит ли удалить пересечение из модели? Почему?
Значимость пересечения определяется по `Pr>f` для `Brand*Technician`. Так как `Pr>f` < 0.0001, то есть меньше 0.05, то пересечение является значимым и удалять его **не стоит**, так как это может привести к некорректным результатам.

#### 3. 
Определите имеются ли различия между специалистами для каждого бренда дисковода. Также исследуйте различия между брендами дисководов для каждого специалиста. 

#### Q5: Каковы ваши выводы? Кому все равно что чинить? Как распределить техников по брендам, чтобы по возможности получить наибольшую производительность труда?
Были получены два графика  **Interaction Plot for Time**: относительно брендов и относительно техников. По графикам видно, что для **Justin** нет разницы что чинить, для остальных техников разница есть. Для более высокой эффективности следует предложить: **Karen** чинить 2-ой бренд, **Bob** 3-ий, **Angela** - 1 и 2, для **Justin** никаких ограничений нет.

```sas
/* код запуска процедуры со всеми опциями, необходимыми для ответов на вопросы в этой работе */
ods graphics on;
proc glm data=work.disks order=internal;
class Brand Technician;
model Time=Brand Technician Brand*Technician;
lsmeans Brand*Technician / slice=Technician;
title 'Analyze the addiction of Service Time from Brand for each Technician';
run;
quit;
```

```sas
/* повторный код запуска процедуры для построения Interaction Plot for Time относительно брендов, а не техников*/
ods graphics on;
proc glm data=work.disks order=internal;
class Technician Brand;
model Time=Technician Brand Technician*Brand;
lsmeans Technician*Brand / slice=Technician;
title 'Analyze the addiction of Service Time from Brand for each Technician';
run;
quit;
```
